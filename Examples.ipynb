```python
%matplotlib inline

import pandas as pd


from bokeh.io import curdoc
from bokeh.themes import Theme
from bokeh import charts 
from bokeh.io import export_png
from IPython.display import Image

size=20
curdoc().theme = Theme(json={'attrs': {

    # apply defaults to Figure properties
    'Title': {
        'text_font_size': '{}pt'.format(size)
    },
    'Axis': {
        'major_label_text_font_size': '{}pt'.format(size*.8),
        'axis_label_text_font_size': '{}pt'.format(size*.8),
    }
}})

charts.defaults.plot_height = 800
charts.defaults.plot_width = 800
```

```python
mpg = pd.read_csv('https://raw.githubusercontent.com/tidyverse/ggplot2/master/data-raw/mpg.csv')
diamonds = pd.read_csv('https://raw.githubusercontent.com/tidyverse/ggplot2/master/data-raw/diamonds.csv')
```

```python
p = charts.Histogram(mpg, 
                     values='cty',
                     bins=12
                    )
Image(export_png(p))
```

```python
p = charts.Scatter(
    data=mpg,
    x='displ',
    y='hwy',
    title="Engine Displacement in Liters vs Highway MPG",
    xlabel='Engine Displacement in Liters',
    ylabel='Highway MPG'
)
Image(export_png(p))
```

```python
p = charts.Scatter(
    mpg,
    x='displ',
    y='hwy',
    color='class',
    title='Engine Displacement in Liters vs Highway MPG',
    xlabel='Engine Displacement in Liters',
    ylabel='Highway MPG'
)
Image(export_png(p))
```
